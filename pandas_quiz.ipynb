### 1. 가장 많이 팔린 아이템은?

```python
import pandas as pd
```

```python
customer_df = pd.read_csv('pandas_data\customer.csv')
customer_df
```

```python
orders_df = pd.read_csv('pandas_data\orders.csv')
orders_df
```

### pandas.Series.idxmax()
- Return the row label of the maximum value.
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.idxmax.html

```python
orders_df.groupby('item').sum()
```

```python
orders_df.groupby('item').sum()['quantity']
```

```python
orders_df.groupby('item').sum()['quantity'].idxmax()
```

### 2. 영희가 가장 많이 구매한 아이템은?

####  1. join
- index를 기준으로 join한다

```python
customer = customer_df.set_index('customer_id')
orders = orders_df.set_index('customer_id')
```

```python
joined_df = customer.join(orders,how = 'outer')
joined_df
```

```python
customer_df.join(orders_df, how = 'outer', on ='customer_id', lsuffix='_left', rsuffix='_right')
```

#### 2. merge
- default : how = inner

```python
merge_df = customer_df.merge(orders_df, on='customer_id', how='outer')
merge_df
```

```python
merge_df.groupby(['name','item']).sum()
```

```python
merge_df.groupby(['name','item']).sum()['quantity']
```

```python
merge_df.groupby(['name','item']).sum()['quantity']['영희']
```

```python
#1
merge_df.groupby(['name','item']).sum()['quantity']['영희'].idxmax()
```

```python
#2
merge_df.groupby(['name','item']).sum().loc['영희','quantity'].idxmax()
```
