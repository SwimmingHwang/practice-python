####  plt.subplot()
- n x n 행렬로 이뤄진 하위 그래프에서 p에 따라 위치를 지정

```python
x = np.arange(0,10,0.1)
y1 = 0.3*(x-5)**2 +1
y2 = -1.5*x +3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1

plt.subplot(2,2,1) # p는 1
plt.plot(x,y1)

plt.subplot(2,2,2) # p는 2
plt.plot(x,y2)

plt.subplot(2,2,3) # p는 3
plt.plot(x,y3)

plt.subplot(2,2,4) # p는 4
plt.plot(x,y4)

plt.show()
```

####  마커의 크기를 인구 밀도에 비례하게 설정

```python
city = ['서울','인천','대전','대구','일산', '울산','부산','광주']

# latitude 위도 longitude 경도
lat = [37.56, 37.45, 36.358, 35.87,35.53,35.18,35.16]
lon = [126.97,126.70,127.38,128.60,129.31,129.07,126.85]

pop_den = [16154,2751,2839,2790,1099,4454,2995]

size = np.array(pop_den)*0.2
colors = ['r','g','b','c','m','k','y'] 

plt.scatter(lon, lat, s=size, c=colors, alpha = 0.5)
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x-0.1, y-0.05, name) # 위도 경도에 맞게 도시 이름 출력 
plt.show()
```

### 막대그래프 (bar)

```python
member_IDs = ['m_01','m_02','m_03','m_04']
before_ex = [27,35,40,33]
after_ex = [30,38,42,37]

n_data = len(member_IDs)
index = np.arange(n_data) # 배열 생성
plt.bar(index, before_ex)
plt.show()
```

- tick_label 옵션

```python
plt.bar(index, before_ex, tick_label = member_IDs)
plt.show()
```

```python
colors = ['r','g','b','m']
plt.bar(index, before_ex,color = colors, tick_label = member_IDs)
plt.show()
```

- index : x축에 표시될 위치, before_ex : height, tick_label:x축 라벨, width : 막대그래프 폭

```python
plt.bar(index, before_ex, tick_label = member_IDs, width = 0.6)
plt.show()
```

```python
colors = ['r','g','b','m']
plt.barh(index, before_ex, color = colors, tick_label = member_IDs)
plt.show()
```

- bar()의 align 옵션은 edge로 지정해 막대 그래프를 한쪽으로 치우치게 함.
- 두개의 데이터를 그린 경우에는 앞에서 이용했던 tick_label옵션을 이용해 tick라벨을 변경할 수 가 없다.   
    이 경우 `xticks(tkck_location, tick_label)`을 이용해 x축의 tick 라벨을 붙임

```python
# aline = center : 막대의 중간지점이 인덱스 값
# aline = edge : 막대의 시작지점(가장자리)가 인덱스값

barWidth = 0.4
plt.bar(index, before_ex, color = 'c', align='edge', width = barWidth, label='before') 
plt.bar(index + barWidth, after_ex, color='m', align = 'edge', width = barWidth, label='after')

plt.xticks(index + barWidth, member_IDs)
plt.legend() # 범례
plt.xlabel('회원 ID')
plt.ylabel('윗몸 일으키기 횟수')
plt.title('운동 시작 전과 후의 근지구력(복근) 변화 비교')
plt.show()
```

### 히스토그램
- 도수분포표를 막대 그래프로 시각화한 것.
- 변량(variable) : 자료를 측정해 숫자로 표시한 것
- 계급(class) : 변량을 정해진 간격으로 나눈 구간
- 도수(frequency) : 나눠진 계급에 속하는 변량의 수 (각 계급에서 발생한 수)
- `plt.hist(x,[,bins-bins_n 혹은 'auto']`

```python
math = [76, 82,84,83,90,86,85,91,71,100,87,81,76,94,78,81,60,79,69,74,87,82,68,79]
plt.hist(math)
```

hist()는 기본적으로 변량을 10개의 계급으로 나눠서 표시함.

```python
plt.hist(math, bins = 8)
plt.show()
```

```python
plt.hist(math, bins = 8)
plt.xlabel('시험 점수')
plt.ylabel('도수(frequency)')
plt.title('수학 시험의 히스토그램')
plt.grid()
plt.show()
```

### 파이그래프
- 파이그래프는 원 안에 데이터의 각 항목이 차지하는 비율만큼 부채꼴의 크기를 갖는 영역으로 이뤄진 그래프
- `plt.pie()`

```python
fruit = ['사과', '바나나','딸기','오렌지','포도']
result = [7,6,3,2,2]
plt.pie(result)
plt.show()
```

`plt.figure(figsize=(w,h))` 
- 그래프의 크기(너비와 높이)를 지정해서 비율을 조절할 수 있음

```python
plt.figure(figsize = (5,5))
plt.pie(result)
plt.show()
```

- `label` 옵션과 `autopct`옵션을 이용해 데이터의 라벨과 비율을 추가할 수 있다

```python
plt.figure(figsize = (5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%')
plt.show()
```

- default는 x축 기준 각도 0도를 시작으로 반시계 방향으로 부채꼴 부분이 그려짐
- x축 기준 각도 90도에서 시작해서 시계방향으로 설정하기 위해 `startangle`옵션과 `counterclock`옵션을 사용하면 된다. 

```python
plt.figure(figsize = (5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle = 90, counterclock = False)
plt.show()
```

- 그림자 추가, 특정요소 강조하기 위해 `shadow`옵션과 `explode`옵션 추가

```python
explode_value = (0.1, 0, 0, 0, 0) # 떨어진 간격

plt.figure(figsize=(5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle = 90, counterclock = False, \
       explode = explode_value, shadow = True)
plt.show()
```

```python
import matplotlib.pyplot as plt

fruit = ['사과', '바나나','딸기','오렌지','포도']
result = [7,6,3,2,2]
explode_value = (0.1, 0, 0, 0, 0) # 떨어진 간격

plt.figure(figsize=(5,5)) # 그래프 크기 지정
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle = 90, counterclock = False, \
       explode = explode_value, shadow = True)

plt.savefig('pandas_data\saveFigTest2.png',dpi=200)
plt.show()
```
