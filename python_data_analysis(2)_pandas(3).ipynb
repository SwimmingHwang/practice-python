### 숫자형 데이터 vs 범주형 데이터 
- 범주형 데이터(Categorical Data) :  몇 개의 범주로 나누어진 자료
    - ex. 성별, 혈액, 성공여부, 효과(없음/조금있음/매우있음) 등
        1. 명목형 자료 (Nominal data) : **성별, 혈액형**과 같이 아무런 순서가 없으며 단순히 분류를 목적으로 함. 
        2. 순서형 자료 (Ordinal data) : **효과**와 같이 명목 척도이면서 '순서'에 의미가 있는 척도
- 숫자형 데이터(Numerical Data)
    - 이산형 자료와 연속형 자료가 있다.   
        1. 이산형 자료 (Discrete Data) : 일정기간 동안의 발생횟수, 출산 횟수와 같은 이산적인 값의 데이터
        2. 연속형 자료 (Continuous Data) : 신장, 체중, 혈압과 같이 연속적인 값을 갖는 데이터

### `df.apply(func)` 함수를 통해 나이데이터 10대,20대(범주형 데이터)로 변환하기

```python
df_train['Age'].head()
```

```python
import math
def age_categorize(age): 
    # cannot convert float NaN to integer 에러때문에 nan처리 해줘야 함.
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
```

```python
df_train['Age'].apply(age_categorize).head()
```

- 함수명로 groupby가능함   
    by : mapping, **function**, label, or list of labels
    Used to determine the groups for the groupby.   
    **If ``by`` is a function**, it's called on **each value of the object's
    index.**

```python
df_train.set_index('Age').groupby(age_categorize).mean()
```

### groupby().Aggregate()

```python
import numpy as np
df_train.groupby(['Pclass','Sex']).aggregate([np.sum, np.mean, np.max, np.min])['Survived']
```

### groupby().transform()
- 평균값을 보여 주지만 모든 row에 대해 해당 그룹 결과값을 보여줌 

```python
df_train.groupby('Pclass').mean()
```

```python
df_train.groupby('Pclass').transform(np.mean).head()
```

- Age2 는 Pclass로 그룹화를 했을 때 해당 그룹의 나이 평균임

```python
df_train['Age2'] = df_train.groupby('Pclass').transform(np.mean)['Age']
```
