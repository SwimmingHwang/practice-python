##### index_col  -  read_csv() 's option
- int or sequence or False, default None
    **Column to use as the row labels of the DataFrame.**   
    If a sequence is given, a
    MultiIndex is used.   
    If you have a malformed file with delimiters at the end
    of each line,   
    you might consider index_col=False to force pandas to _not_
    use the first column as the index (row names)

#### loc()함수를 이용한 인덱싱

````python
df_departments.loc[[10,20,23,26],['DEPARTMENT_ID','MANAGER_ID']]
````

````python
df_departments.iloc[[10,20,23,26],[1,3]]
````

````python
df_departments.iloc[[1,3,5],[1,3]]
````

- iloc : integer position을 통해 값을 찾는다.
- loc : label을 통해서 값을 찾는다.
- 문법 공통점 : df1.loc[[행],[열]]
